```python
with open("data_set.txt", "r", encoding="utf-8") as f:
    lines =  f.readlines()
    lst = []
    for line in lines:
        lst.append(list(map(float, line.split())))

    x = []
    y = []
    for i in range(len(lst) - 1):
        y.append(lst[i+1][1] / lst[i][1] * 100 / (lst[i+1][0] - lst[i][0]))
        x.append((lst[i][0]+lst[i+1][0])/2)

    a = []
    b = []
    for i in range(len(lst)):
        a.append(lst[i][0])
        b.append(lst[i][1]*100/0.62)

    plt.plot(x ,y ,"ro:", a, b, "bo:")
    plt.show()

    print(y[0])
```

```python
n = 100
lx = [i for i in range(n)]
ly = [0.8**i for i in range(n)]

gy = [ly[i+1]/ly[i] for i in range(n - 1)]
gx = [i for i in range(n - 1)]

dy = [ly[i+1] - ly[i] for i in range(n - 1)]
dx = [i for i in range(n - 1)]

plt.plot(lx, ly, "bo-", gx, gy, "ro-", dx, dy, "yo-", [0,100], [0, 0], "g-")
plt.show()
```
